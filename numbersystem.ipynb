Number System

Binary Numbers

```python
a = 25
print(bin(a))  # Output: 0b11001
```

```python
b = 35
print(bin(b))  # Output: 0b100011
k = int(0b100011)  # Output: 35
print(k)
```

```python
print(oct(25))  # Output: '0o31'
print(int(0o31))
```

```python
print(hex(16))  # Output: '0xff'
```

```python
print(0x10)
```

```python
#bitwise operators
x = 16  # In binary: 0b0101
print(-x) # Unary negation
print(~x) # Bitwise NOT: ~0b0101 = 0b1010 (which is -6 in two's complement)
```

```python
#Bitwise AND, OR, XOR
k = 12 & 13
t = 12 | 13
u = 25 ^ 30
print(bin(12))  # Output: 0b1100
print(bin(13))  # Output: 0b1101
print(k)   # Output: 0b1100
print(t)   # Output: 0b1101
print(u)   # Output: 0b0001
```

```python
#Bitwise Leftshift and Rightshift
v = 25 << 1  # Left shift: 0b0101 << 1 = 0b1010 (10 in decimal)
w = 20 >> 1 # Right shift: 0b10100 >> 1 = 0b1010 (10 in decimal)
print(v)  # Output: 10
print(w)  # Output: 10
print(13 >> 3)
print(25 >> 2)
```
