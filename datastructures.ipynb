List indexing and Slicing

```python
l3 = [3,100,2,50]
l4 = l3.copy()
l4.sort(reverse=False)
l5 = l3.copy()
l5.sort(reverse=True)
print(l4)
print(l5)
```

```python
l6 = ['a','z','b','y','c']
print(l6)
l6.sort()
print(l6)
```

```python
l7 = ['a','z','b','y','c']
l7[:]
print(l7[0:3])
print(l7[:-3])
print(l7[3:])
```
