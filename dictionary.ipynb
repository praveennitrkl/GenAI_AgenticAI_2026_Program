Dictionary


```python
d = {1:'one', 2:'two', 3:'three'}
type(d)
print(d.keys())
print(d.values())
print(d.items())
```

```python
d[4] = 'four'
print(d)
d['five'] = 5
print(d)
```

```python
keys ={'a','b'}
dict1 = dict.fromkeys(keys)
print(dict1)
dict2 = dict.fromkeys(keys, 20)
print(dict2)

val = {30,40}
val1 = [30,40]
dict3 = dict.fromkeys(keys, val)
dict4 = dict.fromkeys(keys, val1)
print(dict3)
print(dict4)

print(d)
d.pop(4)
print(d)
```
