Bitwise Operators

```python
a = 5 #the bits of 5 are 0101 in binary
b = 3 #the bits of 3 are 0011 in binary
print(a & b)  # Bitwise AND: 0101 & 0011 = 0001 (1 in decimal)
print(a | b)  # Bitwise OR: 0101 | 0011 = 0111 (7 in decimal)
print(a ^ b)  # Bitwise XOR: 0101 ^ 0011 = 0110 (6 in decimal)
print(~a)     # Bitwise NOT: ~0101 = 1010 (which is -6 in two's complement)
print(a << 1) # Left Shift: 0101 << 1 = 101
print(a >> 1) # Right Shift: 0101 >> 1 = 0010 (2 in decimal)
```

Membership Operators

```python
# Example with the 'in' operator 
fruits = ['apple', 'banana', 'orange', 'guava']
print('apple' in fruits)  # True
print('grape' in fruits)  # False
print('grape' not in fruits)  # True
print('guvava' not in fruits)  # False
```

Identity Operators

```python
x = 10
y = 5
print(x is y)  # False, because x and y are different objects in memory
print(x is not y)  # False, because the values of x and y are different
z = x
print(x is z)  # True, because x and z refer to the same object in memory
print(y is not z)  # True, because y and z refer to different objects in memory
```
