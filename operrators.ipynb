Assignment Operator

```python
x = 2
print(x)
x = x + 2
print(x)
x += 2
print(x)
x -= 2
print(x)
x *= 2
print(x)
x /= 2
print(x)
x //= 2
print(x)
x **= 2
print(x)
print(int(x))
```

Unary Operator

```python
n = 7
print(n)
print(-(n))
```

Relational Operator

```python
a, b = 10, 20
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
print(a == b)
print(a != b)
```

Logical Operator

```python
a = 10
b = 20

print(a < 8 and b > 15)
print(a > 8 and b > 15)
print(a < 8 or b > 15)
print(a > 8 or b > 15)
```
